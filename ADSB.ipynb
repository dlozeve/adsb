```python
cs_table = adsb.loc[:,["hexIdent","callsign"]]
cs_table = cs_table[cs_table.callsign.notnull()]
cs_table = cs_table.sort_values("callsign", axis=0)
cs_table.head(15)
```

```python
positions = adsb[adsb.trans_type == 3]
positions = positions.loc[:,["hexIdent","latitude","longitude","altitude"]]
positions = pd.merge(cs_table, positions, on="hexIdent", how="left")
positions = positions.dropna()
positions.head(15)
```

```python
def df_to_geojson(df, properties, lat='latitude', lon='longitude'):
    geojson = {'type':'FeatureCollection', 'features':[]}
    for _, row in df.iterrows():
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type':'Point',
                               'coordinates':[]}}
        feature['geometry']['coordinates'] = [row[lon],row[lat]]
        for prop in properties:
            feature['properties'][prop] = row[prop]
        geojson['features'].append(feature)
    return geojson
```

```python
geojson = df_to_geojson(positions[0:100], ["hexIdent", "callsign", "altitude"])
```

```python
geojson = {'type':'FeatureCollection', 'features':[]}
for hexIdent in positions['hexIdent'].unique()[1:50]:
    feature = {'type':'Feature',
               'properties':{},
               'geometry':{'type':'LineString',
                           'coordinates':[]}}
    for _, row in positions[positions['hexIdent'] == hexIdent].iterrows():
        feature['geometry']['coordinates'].append([row['longitude'], row['latitude']])
    feature['geometry']['coordinates'].sort()
    for prop in ['hexIdent', 'callsign']:
        feature['properties'][prop] = row[prop]
    geojson['features'].append(feature)
```

```python
geojson_str = json.dumps(geojson, separators=(',', ':'))
```

```python
output_filename = 'aller.geojson'
with open(output_filename, 'w') as output_file:
    output_file.write("var aller = {};".format(geojson_str))
```
